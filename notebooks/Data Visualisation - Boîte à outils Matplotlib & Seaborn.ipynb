# Partie 1 — Data Visualisation

<a id='pourquoi'></a>
## 1.1 Pourquoi visualiser est indispensable en EDA

En 1973, le statisticien Francis Anscombe construit quatre datasets qui ont exactement les mêmes statistiques descriptives : même moyenne, même variance, même corrélation, même droite de régression. Pourtant, tracés sur un graphique, les quatre datasets sont radicalement différents — une relation linéaire, une courbe, une relation linéaire avec un outlier, et des points regroupés sur une verticale. C'est le **Quartet d'Anscombe**.

En 2017, l'histoire se répète avec le **Datasaurus Dozen** : douze datasets avec des statistiques identiques, dont un où les points forment la silhouette d'un dinosaure. Un tableau de chiffres ne le révélera jamais.

Ces deux exemples montrent une vérité fondamentale : **les statistiques résument, la visualisation révèle**. Un résumé statistique peut masquer une relation non linéaire, un cluster, un outlier décisif, ou une structure cachée dans les données. C'est pourquoi en EDA, on visualise avant de modéliser — toujours.

Florence Nightingale l'avait compris en 1858 : son diagramme en roses des vents, montrant que plus de soldats mouraient de maladies évitables que de blessures de guerre, a convaincu le gouvernement britannique de réformer les hôpitaux militaires là où les tableaux de chiffres avaient échoué. **Un bon graphique change des décisions.**

<a id='choisir'></a>
## 1.2 Choisir le bon graphique selon les données

La question à se poser avant de tracer : **qu'est-ce que je veux montrer ?**

**Une variable continue**
- Distribution brute → `hist()` (Matplotlib) / `histplot` (Seaborn)
- Distribution lissée → `kdeplot`
- Distribution + outliers → `boxplot`
- Distribution complète → `violinplot`

**Une variable continue × une catégorie**
- Comparer les distributions → `boxplot` / `violinplot`
- Voir les points individuels → `stripplot` (+ `boxplot`)
- Comparer les moyennes → `barplot` (Seaborn) / `bar()` (Matplotlib)

**Deux variables continues**
- Relation brute → `scatter()` (Matplotlib) / `scatterplot` (Seaborn)
- Relation + régression → `regplot`
- Densité → `kdeplot(x=, y=)`

**Évolution dans le temps ou série ordonnée**
- `plot()` (Matplotlib) / `lineplot` (Seaborn)

**Une variable catégorielle**
- Compter les occurrences → `countplot` / `bar()`
- Proportions → `pie()` (avec parcimonie)

**Toutes les relations d'un coup**
- Vue d'ensemble EDA → `pairplot`
- Corrélations → `heatmap(df.corr())`

**Plusieurs sous-groupes**
- Répéter un graphique → `catplot(col=)` / `displot(col=)`

**Règle générale :** Matplotlib pour le contrôle total et les graphiques sur mesure. Seaborn pour l'EDA rapide et les graphiques statistiques avec intégration DataFrame native.

## 2.1. Le nuage de points `scatter`

Les options de personnalisation :
- la couleur des points, via **color** ou **c** (nom, CSS étendu, HEX, RGB, ou cycle `'C0'`–`'C9'`) ;
- la taille des points, via l'argument **size** ou **s** (standard: 20 à 80, large: 100 à 200; exprimé en pt²);
- le type de marqueur via l'argument **marker** ;
- la transparence des points via l'argument **alpha** (compris entre 0 et 1).

### Choisir un marker sans se perdre

La doc liste 45 markers. En pratique 6-8 utilisé.
3 questions pour filter:

**1. Quel est le rôle sémantique du marker ?**
- Distinguer des catégories → formes géométriques distinctes
- Indiquer un point spécial (outlier, jalon) → marker fort et visible
- Scatter plot dense → petit marker discret, ou pas de marker

**2. Combien de séries à distinguer ?**
- ≤ 4 séries → marker + couleur suffisent
- 5–7 séries → combiner marker + couleur + linestyle
- \> 7 séries → problème de design, pas de marker

**3. Impression N&B ou écran couleur ?**
- Impression N&B → markers différents obligatoires
- Écran couleur → la couleur fait 80% du travail, le marker est secondaire

<a id='pie'></a>
## 2.2 Le diagramme circulaire `pie`

Représente la contribution de chaque catégorie à un total. Pertinent uniquement si les parts s'additionnent à 100% et que les catégories sont peu nombreuses (≤ 6 idéalement).

Arguments principaux :
- **x** : les valeurs numériques agrégées
- **labels** : les étiquettes de chaque part
- **autopct** : pourcentages affichés à l'intérieur des parts — format `'%1.1f%%'`
- **startangle** : angle de départ — `90` pour commencer à 12h
- **explode** : liste de flottants de même longueur que le nombre de parts — détache une part du centre

Pour afficher les pourcentages à l'extérieur, utiliser une légende avec les valeurs calculées manuellement.

> ⚠️ Si une catégorie écrase les autres visuellement, un bar chart horizontal est plus honnête.

````python
data = prets.groupby('ville')['remboursement'].sum()
data = data.reset_index() #pour passer de ville ccomme index à colonne ordinaire
plt.pie(x=data['remboursement'],
        labels=data['ville'],
        autopct='%1.1f%%',
        startangle=90,
        explode=[0,0,0,0,0.1,0]);
````

````python
#Autre option avec une légende pour dégager l'intérieur du diagramme circulaire
data = prets.groupby('ville')['remboursement'].sum().reset_index()

total = data['remboursement'].sum()
labels_legende = [f"{v} ({r/total*100:.1f}%)"
                  for v, r in zip(data['ville'], data['remboursement'])]

plt.pie(x=data['remboursement'], startangle=90)
plt.legend(labels=labels_legende, loc='best');
````

<a id='bar'></a>
## 2.3 Le diagramme en barres `bar` / `barh`

Représente des valeurs numériques par catégorie. À préférer au camembert dès que les catégories dépassent 6 ou qu'une valeur écrase les autres.

Arguments principaux :
- **x** (`bar`) / **y** (`barh`) : les catégories
- **height** (`bar`) / **width** (`barh`) : les valeurs numériques
- **width** (`bar`) / **height** (`barh`) : l'épaisseur des barres — défaut `0.8`
- **color** : couleur des barres — accepte une liste pour colorier chaque barre différemment
- **edgecolor** : couleur du contour des barres
- **bottom** : point de départ vertical — utilisé pour les **barres empilées**

> **`bottom`** : en passant les valeurs de la première série comme `bottom` de la seconde, les barres se superposent proprement.

> **`barh`** : préférer quand les étiquettes sont longues — elles s'affichent horizontalement sans rotation ni chevauchement.



````python
data_sorted = data.sort_values('remboursement', ascending=True)  # Peut trier le data frame en amont pour ordonner la serie
plt.bar(height=data_sorted['remboursement'],
        x=data_sorted['ville'],
        width=0.9,
        color='pink',
        edgecolor='red');
````

````python
data_sorted = data.sort_values('remboursement', ascending=True)
plt.barh(y=data_sorted['ville'],
         width=data_sorted['remboursement'], #on inverse width et height
         height=0.6,
         color='pink',
         edgecolor='red');
````

<a id='hist'></a>
## 2.4 L'histogramme `hist`

Représente la distribution d'une variable numérique continue. Matplotlib découpe automatiquement les valeurs en intervalles (bins) et compte les occurrences dans chacun.

Arguments principaux :
- **x** : la variable numérique à distribuer
- **bins** : nombre d'intervalles — défaut `10`, ajuster selon la densité des données
- **color** : couleur des barres
- **edgecolor** : contour des barres — recommandé pour distinguer les intervalles
- **density** : si `True`, affiche des fréquences relatives (somme = 1) au lieu des effectifs

Arguments de mise en forme (communs à tous les graphiques matplotlib) :
- **plt.title()** : titre du graphique
- **plt.xlabel()** / **plt.ylabel()** : titres des axes
- **plt.xticks()** / **plt.yticks()** : personnalisation des étiquettes d'axes
- **fontsize** : taille de la police
- **fontweight**='bold" : pour mettre en gras


**`bins`** est le paramètre le plus impactant : trop peu de bins écrase la distribution, trop de bins crée du bruit. Tester plusieurs valeurs avant de fixer.


````python
plt.hist(prets['revenu'], bins=30, color='#c9eded', edgecolor='#367d7d')
plt.title('Distribution des revenus', fontsize=18, color='blue')
plt.xlabel('Revenu', fontweight='bold')
plt.ylabel('Effectif');
````

<a id='plot'></a>
## 2.5 Les courbes `plot`

Pour représenter une évolution, typiquement dans le temps.

Arguments principaux :
- **x**, **y** : les coordonnées des points
- **color** : couleur de la courbe
- **linestyle** : style de trait — `'-'` (continu), `'--'` (tirets), `'-.'` (mixte), `':'` (pointillés)
- **linewidth** : épaisseur du trait
- **marker** : symbole aux points de données — voir kit de survie section markers
- **label** : nom de la série — nécessaire pour afficher une légende avec `plt.legend()`

> Quand on trace plusieurs courbes sur le même graphique, `label` + `plt.legend()` deviennent indispensables.

````python
import numpy as np

# Série temporelle synthétique car absente dans le jeu de données utilisé
np.random.seed(42)
mois = np.arange(1, 13)
remboursements = np.cumsum(np.random.randint(1000, 5000, 12)) + 10000

plt.plot(mois, remboursements, marker='o', linestyle='--', color='green',
         linewidth=1.5, label='Remboursements')
plt.title('Évolution mensuelle des remboursements')
plt.xlabel('Mois')
plt.ylabel('Remboursement (€)')
plt.legend();
````

````python
#Exemple avec 2 series
np.random.seed(42)
mois = np.arange(1, 13)
remboursements = np.cumsum(np.random.randint(1000, 5000, 12)) + 10000
objectifs = np.linspace(15000, 50000, 12)  # deuxième série

plt.plot(mois, remboursements, marker='o', linestyle='--', color='green', label='Remboursements')
plt.plot(mois, objectifs, linestyle='-', color='red', label='Objectifs')
plt.title('Évolution mensuelle des remboursements')
plt.xlabel('Mois')
plt.ylabel('Remboursement (€)')
plt.legend();
````

<a id='subplots'></a>
## 2.6 Plusieurs graphiques sur une même fenêtre `subplots`

`plt.subplots()` crée une grille de graphiques indépendants sur une même figure.

Arguments principaux :
- **nrows**, **ncols** : dimensions de la grille
- **figsize** : taille totale de la figure en pouces `(largeur, hauteur)`
- **sharex**, **sharey** : si `True`, impose la même échelle à tous les subplots — indispensable pour comparer honnêtement

`axes.flatten()` transforme la grille 2D en liste simple pour itérer dessus avec `enumerate`.

Sur chaque axe, les méthodes sont les mêmes que `plt.` mais préfixées par `axes[i].` :
- `axes[i].hist()`, `axes[i].scatter()`, `axes[i].plot()`...
- `axes[i].set_title()` : titre du subplot
- `axes[i].set_xlabel()` / `axes[i].set_ylabel()` : titres des axes

`plt.suptitle()` ajoute un titre global au-dessus de tous les subplots — distinct de `set_title()` qui titre chaque subplot individuellement.

`plt.tight_layout()` ajuste automatiquement les espaces entre subplots pour éviter les chevauchements.


````python
villes = prets['ville'].unique()
fig, axes = plt.subplots(2, 3, figsize=(12, 8), sharex=True, sharey=True)
axes = axes.flatten()

for i, ville in enumerate(villes):
    subset = prets[prets['ville'] == ville]
    axes[i].hist(subset['benefices'], bins=20, color='steelblue', edgecolor='white')
    axes[i].set_title(ville)
    axes[i].set_xlabel('Bénéfices')
    axes[i].set_ylabel('Effectif')

plt.suptitle('Distribution des bénéfices par ville', y=1.02)
plt.tight_layout();
````

<a id='perso'></a>
## 2.7 Personnaliser les graphiques

Éléments externes :
- **title** : titre du graphique — accepte `fontsize` et `color`
- **legend** : légende des séries
- **xlabel** / **ylabel** : titres des axes — acceptent `fontsize` et `fontweight`

Éléments internes :
- **ylim** / **xlim** : limites des axes
- **yticks** / **xticks** : graduations manuelles
- **grid** : quadrillage — `axis='y'` pour horizontal uniquement
- **text(x, y, s)** : texte à une position précise du graphique
  - `x` : position horizontale (index de la barre)
  - `y` : position verticale (`hauteur_barre + décalage`)
  - `s` : texte — `round()` pour un entier, f-string pour formatage (`f"{val:,.0f} €"`)
  - kwargs :
     - `ha` :alignement horizontal : left,center, right
     - `va` : alignement vertical : bottom, center, top
     -`fontweight`: bold, light, heavy, black, semibold
     - + `fontsize`, `color`, `rotation`

````python
plt.figure(figsize=(10,6))

plt.bar(height=data_sorted['remboursement'],
        x=data_sorted['ville'],
        width=0.9,
        color='pink',
        edgecolor='red')
plt.title("Chiffres d'affaire réalisé par agence", fontsize=16, color='red')
plt.ylabel("Chiffre d'affaire (€)", fontsize=12, fontweight='bold')
plt.ylim(0, 120000)
plt.yticks([0, 30000, 60000, 90000, 120000])
plt.grid(axis='y')

ca = data_sorted['remboursement'].tolist()

for i in range(len(ca)):
    plt.text(i, ca[i] + 1000, f"{ca[i]:,.0f} €", ha='center')

plt.show()
````

<a id='lignes'></a>
## 2.8 Lignes de référence `axhline` / `axvline`

Tracent une ligne horizontale ou verticale sur un graphique — utile pour marquer une moyenne, un seuil, une valeur cible.

Arguments principaux :
- **y** (axhline) / **x** (axvline) : position de la ligne
- **color**, **linestyle**, **linewidth** : mise en forme
- **label** : pour l'inclure dans la légende



````python

plt.hist(prets['remboursement'], bins=30, color='steelblue', edgecolor='white')
plt.axvline(x=prets['remboursement'].mean(),
            color='red',
            linestyle='--',
            linewidth=1.5,
            label=f"Moyenne : {prets['remboursement'].mean():,.0f} €")
plt.title('Distribution des remboursements')
plt.xlabel('Remboursement (€)')
plt.ylabel('Effectif')
plt.legend()
plt.savefig('distribution_remboursements.png', dpi=150, bbox_inches='tight')
plt.show()
````

<a id='export'></a>
## 2.9 Exporter un graphique `savefig`

À appeler avant `plt.show()` — après, la figure est vidée.

Arguments principaux :
- **fname** : nom du fichier avec extension — `.png`, `.jpg`, `.svg`, `.pdf`
- **dpi** : résolution en points par pouce — `150` pour usage web, `300` pour impression

> **Dans Colab**, `savefig` enregistre le fichier dans la session (pas sur ton ordinateur). Pour le récupérer localement, il faut ensuite le télécharger explicitement. L'ordre est impératif : `savefig` → `show` → `download`.

## 3.1 Configuration / personnlisation

**À faire en début de notebook** pour définir l'apparence globale.

### Thèmes `style` — le fond

Cinq styles disponibles. En pratique deux suffisent pour la majorité des usages :
- `'whitegrid'` — fond blanc avec grille, idéal pour les présentations et rapports
- `'darkgrid'` — fond sombre avec grille, agréable pour l'exploration personnelle

Les autres : `'white'` et `'ticks'` pour les publications, `'dark'` pour un fond sombre sans grille.

### Palettes `palette` — les couleurs

Trois familles :
- **Qualitatives** — pour des catégories distinctes : `'colorblind'` (à privilégier, accessible aux daltoniens), `'deep'` (défaut), `'muted'`
- **Séquentielles** — pour une variable continue du bas vers le haut : `'Blues'`, `'Greens'`, `'viridis'`
- **Divergentes** — pour des valeurs qui s'écartent d'un centre (corrélations, différences) : `'coolwarm'`, `'RdYlGn'`


### Contextes `context` — la taille des éléments

Quatre niveaux du plus petit au plus grand : `'paper'`, `'notebook'` (défaut), `'talk'`, `'poster'`.
**À faire en début de notebook** pour définir l'apparence globale.



````python
# --- Configuration globale ---
sns.set_theme(style='whitegrid', palette='colorblind', context='paper')

# --- Aperçu des palettes qualitatives ---
fig, axes = plt.subplots(3, 1, figsize=(10, 3))
for ax, palette in zip(axes, ['colorblind', 'deep', 'muted']):
    sns.barplot(x=list(range(8)), y=[1]*8, hue=list(range(8)), palette=palette, ax=ax, legend=False)
    ax.set_ylabel(palette, fontsize=10)
    ax.set_xticks([]); ax.set_yticks([])
plt.suptitle('Palettes qualitatives — 8 couleurs')
plt.tight_layout()
plt.show()

# Réinitialiser après l'aperçu
sns.set_theme(style='whitegrid', palette='colorblind');
````

<a id='distrib'></a>
## 3.2 Distributions

Pour visualiser la distribution d'une variable continue, quatre graphiques couvrent la quasi-totalité des besoins :
- `histplot` — le point de départ, distribution brute en barres
- `kdeplot` — version lissée, utile pour comparer plusieurs groupes sur le même graphique
- `boxplot` — résumé en quartiles, excellent pour repérer les outliers et comparer des groupes
- `violinplot` — comme le boxplot mais montre la distribution complète, pas seulement les quartiles

`ecdfplot` existe aussi pour les distributions cumulées, usage plus rare.
<a id='histplot'></a>
### 3.2.1 `histplot` — histogramme

Les paramètres essentiels :
- `x` — variable en abscisse
- `hue` — variable catégorielle pour colorer par groupe
- `bins` — nombre de barres, ou `'auto'`
- `stat` — ce que représente la hauteur des barres : `'count'` (défaut), `'density'`, `'probability'`, `'frequency'`
- `kde=True` — superpose la courbe de densité
- `multiple` — comportement quand `hue` est défini : `'layer'` (superposé, défaut), `'dodge'` (côte à côte), `'stack'`, `'fill'`
- `element` — forme des barres : `'bars'` (défaut), `'step'`, `'poly'`

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Minimal
sns.histplot(data=tips, x='total_bill', ax=axes[0])
axes[0].set_title('Minimal')

# Avec KDE superposée
sns.histplot(data=tips, x='total_bill', kde=True, ax=axes[1])
axes[1].set_title('Avec KDE')

# Par groupe (hue)
sns.histplot(data=tips, x='total_bill', hue='sex',
             multiple='dodge', bins=15, ax=axes[2])
axes[2].set_title('Par sexe (dodge)')

plt.suptitle('histplot — distribution de total_bill')
plt.tight_layout()
plt.show()
````

<a id='kdeplot'></a>
### 3.2.2 `kdeplot` — courbe de densité

Les paramètres essentiels :
- `x` — variable en abscisse ; ajouter `y` pour une KDE bivariée (densité entre deux variables)
- `hue` — une courbe par groupe
- `fill=True` — remplit la surface sous la courbe
- `bw_adjust` — contrôle le lissage : 0.5 donne une courbe fine et détaillée, 2 donne une courbe très lissée
- `common_norm` — `True` normalise tous les groupes ensemble, `False` normalise chaque groupe séparément (utile quand les groupes ont des tailles très différentes)
- `cut` — étend la courbe au-delà des données observées (défaut=3)

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Minimal
sns.kdeplot(data=tips, x='total_bill', ax=axes[0])
axes[0].set_title('Minimal')

# Remplie + par groupe
sns.kdeplot(data=tips, x='total_bill', hue='sex',
            fill=True, alpha=0.5, ax=axes[1])
axes[1].set_title('Remplie, par sexe')

# KDE bivariée (2 variables)
sns.kdeplot(data=tips, x='total_bill', y='tip',
            fill=True, cmap='Blues', ax=axes[2])
axes[2].set_title('Bivariée (total_bill vs tip)')

plt.suptitle('kdeplot — courbe de densité')
plt.tight_layout()
plt.show()
````

<a id='boxplot'></a>
### 3.2.3 `boxplot` — boîte à moustaches

**Lecture du graphique :**
- Trait central = médiane
- Boîte = IQR (Q1 à Q3, 50% des données)
- Moustaches = 1.5 × IQR au-delà des quartiles
- Points au-delà des moustaches = outliers

Les paramètres essentiels :
- `x` — variable catégorielle (les groupes)
- `y` — variable numérique
- `hue` — sous-groupes par couleur
- `order` — liste pour forcer l'ordre des catégories
- `width` — largeur des boîtes (défaut=0.8)
- `fliersize` — taille des points outliers
- `notch=True` — ajoute des encoches représentant l'IC à 95% sur la médiane

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Minimal — une seule variable
sns.boxplot(data=tips, y='total_bill', ax=axes[0])
axes[0].set_title('Minimal (1 variable)')

# Par groupe
sns.boxplot(data=tips, x='day', y='total_bill',
            order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[1])
axes[1].set_title('Par jour (ordre forcé)')

# Par groupe + hue
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex',
            order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[2])
axes[2].set_title('Par jour et sexe')

plt.suptitle('boxplot — distribution par groupe')
plt.tight_layout()
plt.show()
````

<a id='violinplot'></a>
### 3.2.4 `violinplot` — distribution complète

Mêmes paramètres que `boxplot`, avec en plus :
- `inner` — ce qui s'affiche à l'intérieur du violon : `'box'` (défaut), `'quart'`, `'point'`, `'stick'`, `None`
- `split=True` — avec exactement 2 catégories dans `hue`, affiche un côté par groupe sur le même violon
- `bw_adjust` — contrôle le lissage de la courbe, même logique que dans `kdeplot`

**Quand préférer le violin au boxplot ?**
Quand la forme de la distribution compte : une distribution bimodale (deux pics) est invisible dans un boxplot mais immédiatement lisible dans un violin. Même chose pour une asymétrie complexe ou quand on veut montrer la densité complète plutôt qu'un résumé en cinq chiffres.

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Minimal
sns.violinplot(data=tips, x='day', y='total_bill',
               order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[0])
axes[0].set_title('Violin (inner=box)')

# Sans inner
sns.violinplot(data=tips, x='day', y='total_bill',
               order=['Thur', 'Fri', 'Sat', 'Sun'],
               inner='quart', ax=axes[1])
axes[1].set_title('Avec quartiles (inner=quart)')

# Split — un côté par hue (pratique pour comparer 2 groupes)
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex',
               split=True, order=['Thur', 'Fri', 'Sat', 'Sun'],
               inner='quart', ax=axes[2])
axes[2].set_title('Split par sexe')

plt.suptitle('violinplot — distribution complète par groupe')
plt.tight_layout()
plt.show()
````

Trois graphiques couvrent l'essentiel :

### 3.3.1 `scatterplot`

Les paramètres essentiels :
- `x`, `y` — les deux variables continues
- `hue` — couleur par catégorie
- `size` — taille des points proportionnelle à une variable numérique
- `style` — forme des points par catégorie
- `alpha` — transparence entre 0 et 1, utile quand les points se superposent

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Minimal
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[0])
axes[0].set_title('Minimal')

# hue + style
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', style='smoker', ax=axes[1])
axes[1].set_title('Hue + style')

# hue + size (3 dimensions simultanées)
sns.scatterplot(data=penguins.dropna(),
                x='bill_length_mm', y='bill_depth_mm',
                hue='species', size='body_mass_g',
                sizes=(20, 200), alpha=0.7, ax=axes[2])
axes[2].set_title('3 dimensions : hue + size')

plt.suptitle('scatterplot — relations entre variables continues')
plt.tight_layout()
plt.show()
````

<a id='lineplot'></a>
### 3.3.2 `lineplot` — courbes et séries temporelles

Les paramètres essentiels :
- `x`, `y` — les deux variables
- `hue` — une ligne par catégorie
- `estimator` — fonction d'agrégation quand plusieurs valeurs existent pour un même x : `'mean'` (défaut), `'median'`, `'sum'`, `None`
- `errorbar` — bande d'incertitude autour de la ligne : `'ci'` (intervalle de confiance), `'sd'` (écart-type), `'se'` (erreur standard), `None`
- `markers=True` — ajoute des marqueurs sur chaque point
- `dashes=False` — toutes les lignes en trait plein, sans tirets

**À noter :** quand plusieurs observations existent pour un même x, `lineplot` calcule automatiquement la moyenne et affiche un IC à 95% autour de la ligne. Pour désactiver ce comportement : `estimator=None`.

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Série temporelle simple
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
flights_annual = flights.groupby('year')['passengers'].mean().reset_index()
sns.lineplot(data=flights_annual, x='year', y='passengers', ax=axes[0])
axes[0].set_title('Évolution annuelle')

# Par groupe (hue) + IC automatique
# lineplot agrège automatiquement les valeurs par mois
sns.lineplot(data=flights, x='month', y='passengers',
             hue='year', palette='Blues', legend=False, ax=axes[1])
axes[1].set_title('Passagers par mois (une ligne = une année)')

# Avec marqueurs
sns.lineplot(data=flights_annual, x='year', y='passengers',
             markers=True, marker='o', ax=axes[2])
axes[2].set_title('Avec marqueurs')

plt.suptitle('lineplot — évolution et séries')
plt.tight_layout()
plt.show()
````

<a id='regplot'></a>
### 3.3.3 `regplot` et `lmplot` — régression visuelle

Deux fonctions pour le même usage, mais avec une différence importante : `regplot` s'intègre dans une figure existante via `ax=` (interface explicite Matplotlib), tandis que `lmplot` crée sa propre figure et supporte les facettes via `hue` et `col`.

Les paramètres essentiels de `regplot` :
- `x`, `y` — les deux variables
- `ci` — niveau de l'intervalle de confiance en % (95 par défaut, `None` pour le désactiver)
- `scatter=False` — affiche uniquement la droite, sans le nuage de points
- `order` — degré de la régression : 1 (linéaire, défaut), 2 ou plus pour une régression polynomiale
- `logistic=True` — régression logistique quand `y` est binaire (0/1)
- `lowess=True` — régression locale non paramétrique, suit la forme réelle des données sans supposer une relation linéaire

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Régression linéaire avec IC
sns.regplot(data=tips, x='total_bill', y='tip',
            ci=95, ax=axes[0])
axes[0].set_title('Régression linéaire + IC 95%')

# Régression polynomiale (order=2)
sns.regplot(data=tips, x='total_bill', y='tip',
            order=2, ax=axes[1])
axes[1].set_title('Régression polynomiale (order=2)')

# Régression logistique (variable binaire)
titanic_clean = titanic.dropna(subset=['age', 'survived'])
sns.regplot(data=titanic_clean, x='age', y='survived',
            logistic=True, ax=axes[2],
            scatter_kws={'alpha': 0.1})
axes[2].set_title('Régression logistique\n(survie Titanic)')

plt.suptitle('regplot — visualiser une régression')
plt.tight_layout()
plt.show()

# --- lmplot : régression par sous-groupe (facettes) ---
# lmplot crée sa propre figure — ne pas utiliser avec ax=
g = sns.lmplot(data=tips, x='total_bill', y='tip',
               hue='sex', col='time',
               height=4, aspect=0.9)
g.fig.suptitle('lmplot — régression par sexe et moment', y=1.02)
plt.tight_layout()
plt.show()
````

## 3.4. Variables catégorielles

Quatre graphiques couvrent l'essentiel :
- `barplot` — agrège une variable numérique par catégorie (moyenne, somme, médiane)
- `countplot` — compte le nombre d'occurrences par catégorie
- `stripplot` — affiche les points individuels par catégorie

### 3.4.1 `barplot` — agrégation par catégorie

**Important :** `barplot` affiche la moyenne par défaut avec un IC à 95%. Pour un simple comptage, utiliser `countplot`.

Les paramètres essentiels :
- `x` — les catégories
- `y` — la variable numérique à agréger
- `estimator` — fonction d'agrégation : `'mean'` (défaut), `'sum'`, `'median'`, `'std'`
- `errorbar` — barre d'erreur : `'ci'` (IC), `'sd'`, `'se'`, `None`
- `hue` — sous-groupes par couleur
- `order` — liste pour forcer l'ordre des catégories
- `orient` — `'v'` (vertical, défaut) ou `'h'` (horizontal)

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Moyenne par jour (défaut)
sns.barplot(data=tips, x='day', y='total_bill',
            order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[0])
axes[0].set_title('Moyenne par jour + IC 95%')

# Somme par jour
sns.barplot(data=tips, x='day', y='total_bill',
            estimator='sum', errorbar=None,
            order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[1])
axes[1].set_title('Somme par jour (sans IC)')

# Horizontal + hue
sns.barplot(data=tips, y='day', x='total_bill', hue='sex',
            orient='h', order=['Thur', 'Fri', 'Sat', 'Sun'],
            ax=axes[2])
axes[2].set_title('Horizontal + hue')

plt.suptitle('barplot — agrégation par catégorie')
plt.tight_layout()
plt.show()
````

<a id='countplot'></a>
### 3.4.2 `countplot` — compter les occurrences

Les paramètres essentiels :
- `x` ou `y` — la variable catégorielle à compter (utiliser `y` pour un graphique horizontal)
- `hue` — sous-groupes par couleur
- `order` — liste pour forcer l'ordre des catégories, ou `tips['day'].value_counts().index` pour trier par fréquence décroissante
- `stat` — `'count'` (défaut) ou `'percent'` pour afficher des pourcentages

**À noter :** `countplot` est équivalent à `histplot` avec `discrete=True`. Les deux produisent le même résultat — `countplot` est simplement plus explicite pour les variables catégorielles.

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Minimal
sns.countplot(data=tips, x='day',
              order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[0])
axes[0].set_title('Nombre de repas par jour')

# Avec pourcentage
sns.countplot(data=tips, x='day', stat='percent',
              order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[1])
axes[1].set_ylabel('%')
axes[1].set_title('Pourcentage par jour')

# Trié par fréquence décroissante
order_by_freq = tips['day'].value_counts().index
sns.countplot(data=tips, x='day', hue='sex',
              order=order_by_freq, ax=axes[2])
axes[2].set_title('Trié par fréquence + hue')

plt.suptitle('countplot — comptage par catégorie')
plt.tight_layout()
plt.show()
````

<a id='stripplot'></a>
### 3.4.3 `stripplot` — points individuels par catégorie

Les paramètres essentiels :
- `x`, `y` — la catégorie et la variable numérique
- `hue` — couleur par sous-groupe
- `jitter=True` — disperse légèrement les points horizontalement pour éviter la superposition (défaut activé)
- `alpha` — transparence, utile quand les points sont nombreux
- `size` — taille des points

`swarmplot` fonctionne de façon identique mais garantit l'absence de superposition en positionnant chaque point précisément — plus lisible mais nettement plus lent sur les grands datasets.

**Bonne pratique :** superposer `stripplot` sur un `boxplot` ou un `violinplot` pour voir à la fois le résumé statistique et les points individuels. Appeler d'abord le boxplot, puis le stripplot sur le même axe.

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Boxplot + stripplot superposés
sns.boxplot(data=tips, x='day', y='total_bill',
            order=['Thur', 'Fri', 'Sat', 'Sun'],
            color='lightblue', ax=axes[0])
sns.stripplot(data=tips, x='day', y='total_bill',
              order=['Thur', 'Fri', 'Sat', 'Sun'],
              color='darkblue', alpha=0.4, size=4, ax=axes[0])
axes[0].set_title('Boxplot + stripplot\n(distribution + points individuels)')

# Violinplot + stripplot superposés
sns.violinplot(data=tips, x='day', y='total_bill',
               order=['Thur', 'Fri', 'Sat', 'Sun'],
               inner=None, color='lightgreen', ax=axes[1])
sns.stripplot(data=tips, x='day', y='total_bill',
              order=['Thur', 'Fri', 'Sat', 'Sun'],
              color='darkgreen', alpha=0.4, size=3, ax=axes[1])
axes[1].set_title('Violinplot + stripplot')

plt.suptitle('Superposition — distribution + points individuels')
plt.tight_layout()
plt.show()
````

## 3.5. Matrices

### 3.5.1 `heatmap` — matrice de corrélation et tableaux croisés

Les paramètres essentiels :
- `data` — un DataFrame carré, typiquement `df.corr()` pour une matrice de corrélation
- `annot=True` — affiche les valeurs numériques dans chaque case
- `fmt` — format des valeurs affichées : `'.2f'` pour deux décimales, `'d'` pour des entiers
- `cmap` — palette : `'coolwarm'` pour les corrélations (divergente autour de 0), `'Blues'` pour des fréquences
- `vmin`, `vmax` — limites de l'échelle de couleur, à fixer à -1 et 1 pour une matrice de corrélation
- `center` — valeur centrale de la palette, mettre 0 pour une corrélation
- `linewidths` — épaisseur de la séparation entre les cases
- `mask` — masque booléen pour cacher certaines cases, typiquement le triangle supérieur pour éviter la redondance d'une matrice de corrélation

````python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Matrice de corrélation
corr = tips.select_dtypes('number').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            vmin=-1, vmax=1, center=0, ax=axes[0])
axes[0].set_title('Matrice de corrélation')

# Triangle inférieur seulement (évite la redondance)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            vmin=-1, vmax=1, center=0, mask=mask, ax=axes[1])
axes[1].set_title('Triangle inférieur\n(sans redondance)')

# Tableau croisé (contingence)
ct = pd.crosstab(tips['day'], tips['time'],
                 values=tips['total_bill'], aggfunc='mean')
sns.heatmap(ct, annot=True, fmt='.1f', cmap='YlOrRd',
            linewidths=0.5, ax=axes[2])
axes[2].set_title('Addition moyenne\npar jour et moment')

plt.suptitle('heatmap — matrices et tableaux croisés')
plt.tight_layout()
plt.show()
````

`pairplot` est l'outil EDA numéro un en début d'exploration : une seule ligne de code produit une grille complète qui montre toutes les relations entre variables numériques et les distributions sur la diagonale. Avec `hue`, chaque groupe apparaît dans une couleur distincte — les clusters et séparations deviennent immédiatement visibles.

Les paramètres essentiels :

**Quand utiliser :** comparer la même visualisation sur plusieurs sous-groupes.

# Partie 4 - Combiner Seaborn et Matplotlib

Seaborn génère des graphiques propres par défaut. Pour aller plus loin — titres, labels d'axes, annotations, légende personnalisée — on récupère l'axe et on utilise les fonctions Matplotlib dessus. Les deux librairies se complètent naturellement : Seaborn trace, Matplotlib affine.

La clé est de toujours passer `ax=` à la fonction Seaborn pour contrôler sur quel axe elle dessine, puis d'appeler les méthodes Matplotlib sur cet axe.

````python
fig, ax = plt.subplots(figsize=(9, 5))

# Graphique Seaborn
sns.barplot(data=tips, x='day', y='total_bill', hue='sex',
            order=['Thur', 'Fri', 'Sat', 'Sun'],
            errorbar='se', ax=ax)

# Personnalisation Matplotlib sur l'axe récupéré
ax.set_title('Addition moyenne par jour et sexe\n(barres d\'erreur = erreur standard)',
             fontsize=13, pad=12)
ax.set_xlabel('Jour de la semaine', fontsize=11)
ax.set_ylabel('Addition moyenne ($)', fontsize=11)
ax.set_xticklabels(['Jeudi', 'Vendredi', 'Samedi', 'Dimanche'])
ax.legend(title='Sexe', labels=['Femme', 'Homme'])
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)  # grille derrière les barres

# Annotation sur une barre
ax.annotate('Week-end\npic', xy=(2, 20), fontsize=9,
            color='gray', ha='center')

sns.despine()  # supprimer les bordures droite et haut
plt.tight_layout()
plt.show()
````

````python
fig = plt.figure(figsize=(16, 10))

# Disposition avec gridspec pour contrôle précis
gs = fig.add_gridspec(2, 3, hspace=0.4, wspace=0.35)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])
ax4 = fig.add_subplot(gs[1, 0])
ax5 = fig.add_subplot(gs[1, 1])
ax6 = fig.add_subplot(gs[1, 2])

# 1. Distribution de la variable cible
sns.histplot(data=tips, x='total_bill', kde=True, ax=ax1)
ax1.set_title('Distribution — addition')

# 2. Corrélation numérique
corr = tips.select_dtypes('number').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            vmin=-1, vmax=1, center=0, ax=ax2)
ax2.set_title('Corrélations')

# 3. Distribution par groupe
sns.boxplot(data=tips, x='day', y='total_bill',
            order=['Thur', 'Fri', 'Sat', 'Sun'], ax=ax3)
ax3.set_title('Addition par jour')

# 4. Relation continue
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', alpha=0.7, ax=ax4)
ax4.set_title('Addition vs pourboire')

# 5. Comptage catégoriel
sns.countplot(data=tips, x='day', hue='sex',
              order=['Thur', 'Fri', 'Sat', 'Sun'], ax=ax5)
ax5.set_title('Nombre de repas par jour')

# 6. Distribution par 2 catégories
sns.violinplot(data=tips, x='time', y='total_bill',
               hue='sex', split=True, ax=ax6)
ax6.set_title('Addition par moment et sexe')

fig.suptitle('EDA — Dataset Tips', fontsize=15, fontweight='bold', y=1.01)
plt.show()
````
