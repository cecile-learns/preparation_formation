## 1 - Multiplication scalaire et somme de vecteurs

### 1.1 - Visualiation d'un vecteur $v\in\mathbb{R}^2$

Un vecteur a une magnitude ou norme, une direction et un sens. Ils peuvent être visualisé sous forme d'une flèche. Exemple pris :

pour $v\in\mathbb{R}^2$, $v=\begin{bmatrix}
 1 & 3
\end{bmatrix}^T$



```python
def visu_vecteur(list_v, list_label, list_color):
    _, ax = plt.subplots(figsize=(10, 10))
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.set_xticks(np.arange(-10, 10))
    ax.set_yticks(np.arange(-10, 10))

    plt.axis([-10, 10, -10, 10])
    for i, v in enumerate(list_v):
        sgn = 0.4 * np.array([[1] if i==0 else [i] for i in np.sign(v)])
        plt.quiver(v[0], v[1], color=list_color[i], angles='xy', scale_units='xy', scale=1)
        ax.text(v[0]-0.2+sgn[0], v[1]-0.2+sgn[1], list_label[i], fontsize=14, color=list_color[i])

    plt.grid()
    ax.axhline(y=0, color='k', linewidth=1.5)
    ax.axvline(x=0, color='k', linewidth=1.5)
    plt.gca().set_aspect("equal")
    plt.show()
```

```python
v=np.array([[1],[3]]) # Utilisation d'un shape (2,1) âr convention mathématique
visu_vecteur([v],[f"$v$"], ["steelblue"] )
```

### 1.2 - Multiplication par un scalaire

Soit v un vecteur, $v=\begin{bmatrix}
 v_1 & v_2 & \ldots & v_n
\end{bmatrix}^T\in\mathbb{R}^n$

Soit $k\in\mathbb{R}$

$kv=\begin{bmatrix}
 kv_1 & kv_2 & \ldots & kv_n
\end{bmatrix}^T$


*   k>1 : étirement
*   0<k<1 : rétrécissement
*   k=0 : vecteur nul
*   k<0 : inversion de direction








```python
visu_vecteur([v, 2*v, 0.5*v,-2*v],
 [f"$v$", f"$2v$", f"$0,5v$",f"$-2v$"],
 ["black", "green", "blue","red"])
```

<a name='1.3'></a>

### 1.3 - Somme des vecteurs

La somme de vecteurs s'effectue en additionnant les composants correspondants.

Relation de Chalsles : [A,C] = [A,B] + [B,C]



```python
v=np.array([[1],[3]])
w=np.array([[4],[-1]])

visu_vecteur([v, w, v+w], [f"$v$", f"$w$", f"$v + w$"], ["black", "black", "red"])
```

### 2.1 - Définition algébrique du produit scalaire

Le **produit scalaire** est la somme des produits de leurs composantes correspondantes : un nombre qui encode à la fois les normes et l'angle entre les vecteurs.

$x=\begin{bmatrix}
 x_1 & x_2 & \ldots & x_n
\end{bmatrix}^T\in\mathbb{R}^n$ and  
$y=\begin{bmatrix}
 y_1 & y_2 & \ldots & y_n
\end{bmatrix}^T\in\mathbb{R}^n$

$$x\cdot y = \sum_{i=1}^{n} x_iy_i = x_1y_1+x_2y_2+\ldots+x_ny_n \tag{1}$$



```python
#exemple 1 avec shape(2,1):
v=np.array([[1],[3]])
w=np.array([[4],[-1]])

print(f"Le produit scalaire de v et w est : {np.dot(v.flatten(), w.flatten())}")
```

```python
#exemple 2 avec shape(2, ):
v=np.array([1,3])
w=np.array([4,-1])

print(f"Le produit scalaire de v et w est : {v@w}")
```

### 2.2 - Définition géométrique du produit scalaire

Le produit scalaire de u et v vaut ‖u‖ × ‖v‖ × cos(θ), où θ est l'angle entre eux.

- θ =0° : les vecteurs sont alignés , 100% de projection
- θ =90° : les vecteurs sont perpendiculaire, 0% de projection
- θ=180° : les vecteurs sont opposés
- θ=60 alors cos(θ)=0,5, il y a 50% de projection
- θ=120° alors cos(θ)=-0,5, il a -50% de projection

autrement dit
- u.v > 0 angle aigue, direction similaire
- u.v=0 : angle droit, perpendiculaire
- u.v<0 : angle obtus : direction opposées



```python
# Illustration avec Claude Ai
fig, ax = plt.subplots(figsize=(8, 8))

# Vecteur u incliné
u = np.array([3, 3])

# Projection de u sur l'axe x (v = [1, 0])
v = np.array([1, 0])
proj_vecteur = np.array([u[0], 0])  # projection sur x = [ux, 0]

# Calculs
ps = np.dot(u, v)
cos_theta = ps / (np.linalg.norm(u) * np.linalg.norm(v))
theta = np.degrees(np.arccos(np.clip(cos_theta, -1, 1)))

# Vecteur u
ax.quiver(0, 0, u[0], u[1], color="black", angles='xy', scale_units='xy', scale=1)
ax.text(u[0]+0.1, u[1]+0.1, "$u$", fontsize=14, color="black")

# Axe x comme vecteur v
ax.quiver(0, 0, 5, 0, color="red", angles='xy', scale_units='xy', scale=1)
ax.text(5.1, 0.1, "$v$", fontsize=14, color="red")

# Projection (flèche bleue sur l'axe x)
ax.quiver(0, 0, proj_vecteur[0], proj_vecteur[1], color="blue", angles='xy', scale_units='xy', scale=1)
ax.text(proj_vecteur[0]/2, -0.3, f"‖u‖ × cos(θ) = {np.linalg.norm(u)*cos_theta:.2f}", fontsize=11, color="blue", ha='center')

# Ligne perpendiculaire (tiret vert)
ax.plot([u[0], proj_vecteur[0]], [u[1], proj_vecteur[1]], 'g--', linewidth=1.5, label="perpendiculaire")

# Arc pour l'angle
arc = plt.matplotlib.patches.Arc((0, 0), 1.5, 1.5, angle=0, theta1=0, theta2=theta, color="purple", linewidth=1.5)
ax.add_patch(arc)
ax.text(1.0, 0.3, f"θ={theta:.0f}°", fontsize=12, color="purple")

# Annotation u.v
ax.text(1.5, 2.5, f"u · v = ‖u‖ × ‖v‖ × cos(θ)\n= {np.linalg.norm(u):.2f} × {np.linalg.norm(v):.2f} × {cos_theta:.2f}\n= {ps:.2f}",
        fontsize=11, color="black",
        bbox=dict(boxstyle="round", facecolor="lightyellow", alpha=0.8))

# Mise en forme
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 5)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
ax.legend(fontsize=10)
ax.set_title("Définition géométrique du produit scalaire\nu · v = ‖u‖ × ‖v‖ × cos(θ)", fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 2.3 - Projection d'un vecteur sur un autre

La projection de u sur v répond à : quelle est l'"ombre" de u
dans la direction de v ?

Formule : proj = (u·v / ‖v‖²) × v

En traitement du langage naturel, on transforme les mots en vecteurs dans un espace multidimensionnel. Des mots sémantiquement proche autont des vecteurs proches et donc des angles proche de 0. C'est le cas de chien et loup. Amour et haine seront opposées et chien et ordi seront perpendiculaire car rien en commun.

```python
# Illustration avec Claude Ai
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

cas = [
    {"u": np.array([1, 0]),  "v": np.array([1, 0]),   "titre": "θ=0° — Alignés (100% projection)",      "couleurs": ["black", "red"]},
    {"u": np.array([1, 0]),  "v": np.array([0, 1]),   "titre": "θ=90° — Perpendiculaires (0% projection)", "couleurs": ["black", "red"]},
    {"u": np.array([1, 0]),  "v": np.array([-1, 0]),  "titre": "θ=180° — Opposés (-100% projection)",   "couleurs": ["black", "red"]},
    {"u": np.array([3, 4]),  "v": np.array([4, 3]),   "titre": "Cas ML — Profils similaires",            "couleurs": ["black", "red"]},
]

for ax, c in zip(axes.flatten(), cas):
    u, v = c["u"], c["v"]

    # Flèches
    ax.quiver(0, 0, u[0], u[1], color=c["couleurs"][0], angles='xy', scale_units='xy', scale=1, label="u")
    ax.quiver(0, 0, v[0], v[1], color=c["couleurs"][1], angles='xy', scale_units='xy', scale=1, label="v")

    # Projection de u sur v
    v_norm = v / np.linalg.norm(v)
    proj_scalaire = np.dot(u, v_norm)
    proj_vecteur = proj_scalaire * v_norm
    ax.quiver(0, 0, proj_vecteur[0], proj_vecteur[1], color="blue", angles='xy',
          scale_units='xy', scale=1, alpha=0.5, label="proj")
    ax.plot([u[0], proj_vecteur[0]], [u[1], proj_vecteur[1]],
            'g--', alpha=0.5, label="distance")

    # Calculs
    ps = np.dot(u, v)
    cos_theta = ps / (np.linalg.norm(u) * np.linalg.norm(v))
    theta = np.degrees(np.arccos(np.clip(cos_theta, -1, 1)))

    # Mise en forme
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.set_title(c["titre"], fontsize=11, fontweight='bold')
    ax.legend(loc='upper left', fontsize=9)
    ax.set_xlabel(f"u·v = {ps:.2f} | θ = {theta:.1f}° | cos(θ) = {cos_theta:.2f}", fontsize=9)

plt.suptitle("Définition géométrique du produit scalaire", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
